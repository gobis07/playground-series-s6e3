```python
train_data = train_data.drop('id',axis = 1)
```

PRETRAITEMENT DE DONNEES

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
import numpy as np
le = LabelEncoder()
scaler = StandardScaler()
def encoder(df):
     df['gender'] = df['gender'].map({'Male':0,'Female':1})
     df['Partner'] = df['Partner'].map({'No':0, 'Yes':1})
     df['Dependents'] = df['Dependents'].map({'No':0, 'Yes':1})
     df['PhoneService'] = df['PhoneService'].map({'No':0, 'Yes':1})
     df['PaperlessBilling'] = df['PaperlessBilling'].map({'No':0, 'Yes':1})
     df['MultipleLines'] = df['MultipleLines'].replace('No phone service', 'No')
     df['OnlineSecurity'] = df['OnlineSecurity'].replace('No internet service', 'No')
     df['OnlineBackup'] = df['OnlineBackup'].replace('No internet service', 'No')
     df['DeviceProtection'] = df['DeviceProtection'].replace('No internet service', 'No')
     df['TechSupport'] = df['TechSupport'].replace('No internet service', 'No')
     df['StreamingTV'] = df['StreamingTV'].replace('No internet service', 'No')
     df['StreamingMovies'] = df['StreamingMovies'].replace('No internet service', 'No')
     df['MultipleLines'] = df['MultipleLines'].map({'No':0, 'Yes':1})
     df['OnlineSecurity'] = df['OnlineSecurity'].map({'No':0, 'Yes':1})
     df['OnlineBackup'] = df['OnlineBackup'].map({'No':0, 'Yes':1})
     df['DeviceProtection'] = df['DeviceProtection'].map({'No':0, 'Yes':1})
     df['TechSupport'] = df['TechSupport'].map({'No':0, 'Yes':1})
     df['StreamingTV'] = df['StreamingTV'].map({'No':0, 'Yes':1})
     df['StreamingMovies'] = df['StreamingMovies'].map({'No':0, 'Yes':1})
     df['InternetService'] = df['InternetService'].map({'No':0, 'DSL':1, 'Fiber optic':2})
     df['Contract'] = df['Contract'].map({'Month-to-month':0, 'One year':1, 'Two year':2})
     df['PaymentMethod'] = df['PaymentMethod'].map({'Mailed check':0,
                                                'Electronic check':1,
                                                'Bank transfer (automatic)':2,
                                                'Credit card (automatic)':3})
     df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
     train_data['Churn'] = train_data['Churn'].map({'No':0,'Yes':1})
     train_data['MonthlyCharges'] = np.log(train_data['MonthlyCharges'])
     train_data['tenure'] = np.log(train_data['tenure'])
     return df

train_data_encoder = encoder(train_data)
def feature_engg(df):
    df['ChargesPerDay'] = (df['MonthlyCharges']/30).round(3)
    df['ChargesPerHours'] = (df['ChargesPerDay']/24).round(3)
    df['SeniorCitizensWithPartner'] = ((df['SeniorCitizen']== 1) & (df['Partner'] == 1)).astype('int')
    return df
train_data_engg = feature_engg(train_data_encoder)
```

ENTRAINEMENT DU MODEL

```python
from sklearn.model_selection import train_test_split
x = train_data_engg.drop(['Churn'],axis = 1)
y = train_data_engg['Churn']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,stratify=y,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
rfc = RandomForestClassifier(
                             n_estimators=100,
                             min_samples_split=3,
                             max_depth=20,
                             min_samples_leaf = 1,
                             criterion = 'entropy', 
                             max_features = 'log2',
                             class_weight= 'balanced'
                             )
xgb = XGBClassifier(
                    n_estimators = 200,
                    max_depth = 8,
                    learning_rate = 0.1
)
LGBM = LGBMClassifier(
                    max_depth=3, n_estimators=200, learning_rate=0.05,
                    device='gpu', random_state=42, verbose=-1
    )
LGBM.fit(x_train,y_train)
xgb.fit(x_train,y_train)
rfc.fit(x_train,y_train)
y_proba_lgbm = LGBM.predict_proba(x_test)[:,1]
y_proba_xgb = xgb.predict_proba(x_test)[:,1]
y_proba_rfc = rfc.predict_proba(x_test)[:,1]
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

stack = StackingClassifier(
    estimators=[
        ('rfc',rfc),
        ('xgb',xgb),
        ('lgbm',LGBM)
    ],
    final_estimator=LogisticRegression()
)
stack.fit(x_train,y_train)
y_proba_stack = stack.predict_proba(x_test)[:,1]
```

TUNNING DES HYPERPARAMETRES

```python
from sklearn.model_selection import RandomizedSearchCV

params = {
   'n_estimators':[100,200],
   'max_depth':[6,8],
   'learning_rate':[0.01,0.1],

}
rand = RandomizedSearchCV(xgb,
                          param_distributions=params,
                          n_iter=4,
                          cv = 3,
                          n_jobs=1,
                          scoring = 'roc_auc')
rand.fit(x_train,y_train)
print('best paramètre',rand.best_params_)
print('best score',rand.best_score_)
```

EVALUATION

```python
from sklearn.metrics import roc_auc_score

print('roc_auc_rfc',roc_auc_score(y_test,y_proba_rfc))
print('roc_auc_lgbm',roc_auc_score(y_test,y_proba_lgbm))
print('roc_auc_xgb',roc_auc_score(y_test,y_proba_xgb))
print('roc_auc_stacking',roc_auc_score(y_test,y_proba_stack))
```

```python
from sklearn.model_selection import cross_val_score

scores_rfc = cross_val_score(rfc,x,y,cv=5,scoring='roc_auc')
scores_xgb = cross_val_score(xgb,x,y,cv=5,scoring='roc_auc')
print('rfc :',np.mean(scores_rfc))
print('xgb :',np.mean(scores_xgb))
```

VISUALISATION DU COURBE D'APPRENTISSAGE

```python
import matplotlib.pyplot as plt
import numpy as np
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import LearningCurveDisplay, ShuffleSplit

fig, ax = plt.subplots(1,3,figsize = (15,5))
models = {
    'XGBoost' : XGBClassifier(
        n_estimators = 200,
        max_depth = 8,
        learning_rate = 0.1
    ),
    'LightGBM' :LGBMClassifier(
        max_depth=100,
        learning_rate=0.1
    ),
    'RandomForest':RandomForestClassifier(
        n_estimators=100,
        min_samples_split=3,
        max_depth=10,
        min_samples_leaf = 1,
        criterion = 'entropy', 
        max_features = 'log2',
        class_weight= 'balanced'
    )
}
cv = ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)

for i,(name,model) in enumerate(models.items()):
    LearningCurveDisplay.from_estimator(model,
                                        x,
                                        y,
                                        cv=cv,
                                        train_sizes=np.linspace(0.1,1.0,5))
    ax[i].set_title(name)
plt.show()
```
